## 1. Synthetic Data Generation

### What's Being Simulated
We create synthetic cross-sectional images that simulate internal structures:
- **Outer Shell**: High-density outer wall (like a fruit skin)
- **Inner Tissue**: Medium-density flesh material
- **Core Structure**: Optional central seed/core
- **Defects**: Random anomalies that reduce quality

```python
# Initialize the data generator
generator = SyntheticMRIGenerator(image_size=256)

# Generate a sample with seed for reproducibility
img_controlled, meta_controlled = generator.generate_sample(
    seed=42,
    has_core=True,      # Force a core
    has_defect=False    # No defects (should be Premium)
)

print("✓ Synthetic MRI sample generated!")
print(f"  Image shape: {img_controlled.shape}")
print(f"  Intensity range: [{img_controlled.min():.3f}, {img_controlled.max():.3f}]")
```

```python
# Visualize the synthetic image
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Original image
axes[0].imshow(img_controlled, cmap='gray')
axes[0].set_title('Synthetic MRI Scan', fontweight='bold')
axes[0].axis('off')

# Histogram
axes[1].hist(img_controlled.flatten(), bins=50, color='steelblue', edgecolor='black', alpha=0.7)
axes[1].set_xlabel('Intensity Value')
axes[1].set_ylabel('Frequency')
axes[1].set_title('Intensity Distribution', fontweight='bold')
axes[1].grid(True, alpha=0.3)

# Radial profile
center = 128
radius_line = img_controlled[center, :]
axes[2].plot(radius_line, linewidth=2, color='darkgreen')
axes[2].fill_between(range(len(radius_line)), radius_line, alpha=0.3)
axes[2].set_xlabel('Distance from Center')
axes[2].set_ylabel('Intensity')
axes[2].set_title('Radial Intensity Profile', fontweight='bold')
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\nImage Characteristics:")
print(f"  Mean intensity: {img_controlled.mean():.3f}")
print(f"  Std deviation: {img_controlled.std():.3f}")
print(f"  Dynamic range: {img_controlled.max() - img_controlled.min():.3f}")
```

## 2. Feature Extraction

### Interpretable Features (Not ML Black Box)
Instead of relying on learned representations, we compute **structural measurements**:

| Feature Category | Examples | Purpose |
|------------------|----------|----------|
| **Intensity Stats** | mean, std, range | Overall brightness and variation |
| **Density Distribution** | radial profile, gradient | Shell-to-center density flow |
| **Layer Detection** | wall thickness, layer count | Structural integrity |
| **Symmetry** | horizontal, vertical, overall | Uniformity and shape |
| **Anomaly Detection** | presence, count, severity | Defect identification |
| **Uniformity** | variance-based score | Material consistency |

```python
# Initialize feature extractor
extractor = FeatureExtractor()

# Extract features
features = extractor.extract_features(img_controlled)

# Print human-readable feature summary
extractor.print_features(features)

print("\n✓ All features extracted successfully!")
```

---

## 3. Rule-Based Classification

### Classification Logic
Quality is determined by **transparent threshold-based rules**:

- **Rule 1**: Check for anomalies → IF severity > 0.02 THEN -50 pts
- **Rule 2**: Evaluate uniformity → IF score < 0.60 THEN -20 pts
- **Rule 3**: Check symmetry → IF score < 0.65 THEN -15 pts
- **Rule 4**: Validate wall thickness → IF outside [0.05, 0.25] THEN -10/15 pts
- **Rule 5**: Bonus for core → IF well-formed THEN +5 pts

**Final Classifications**:
- Score ≥ 85 → **Premium** ✅
- Score ≥ 60 → **Standard** ⚠️
- Score < 60 → **Defective** ❌

```python
# Initialize the rule-based classifier
classifier = QualityClassifier()

# Show current thresholds
print("Classification Thresholds:")
for key, value in classifier.thresholds.items():
    print(f"  {key}: {value}")
```

```python
# Classify the sample
classification = classifier.classify(features)

# Display results
classifier.print_classification(classification)
```
